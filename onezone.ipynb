```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from onezone import *
```

```python
onezone = OneZone()
```

Run the one zone model with accretion rate 1.0 Eddington for $10^4$ seconds

```python
time, y, T, L, deltat = onezone.evolve(1.0, 1e4)
```

```python
plt.plot(time,L)
plt.xlabel('Time (s)')
plt.ylabel('Luminosity (erg/s)')
```

```python
plt.plot(y,T)
plt.xlabel('Column depth (g/cm^2)')
plt.ylabel('Temperature (K)')
```
